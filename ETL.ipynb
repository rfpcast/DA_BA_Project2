```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1_ETL = pd.read_excel('Data\homicidios.xlsx')
```

```python
df_1_ETL_v2 = df_1_ETL.copy()
```

### Normalizacion

```python
df_1_ETL_v3 = df_1_ETL_v2.copy()
```

```python
df_ver1 = df_1_ETL_v3[df_1_ETL_v3["Cruce"].isnull()]
df_ver2 = df_1_ETL_v3[df_1_ETL_v3["Calle"].isnull()]
df_ver3 = df_1_ETL_v3[df_1_ETL_v3["Dirección Normalizada"].isnull()]
```

Columna **"Cruce"**

```python
df_ver1.head(5)
```

```python
df_1_ETL_v3["Cruce"].replace(to_replace=np.nan, value="Sin Cruce", inplace=True)
```

```python
df_1_ETL_v3["Cruce"].isnull().sum()
```

```python
indices_df_ver1 = df_ver1.index

seleccionados = df_1_ETL_v3.loc[indices_df_ver1]
```

```python
seleccionados.head(5)
```

```python
df_ver3.head()
```
